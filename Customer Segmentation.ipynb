# Project Customer Segmentation


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('D:\CODE\Jupyter Notebook\Project\Customer Segmentation\Mall_Customers.csv')
```

## Univariate Analysis

```python
df.head()
```

```python
df.describe()
```

```python
columns = ['Age', 'Annual Income (k$)','Spending Score (1-100)']
for i in columns:
    plt.figure()
    sns.histplot(df[i])
```

```python
columns = ['Age', 'Annual Income (k$)','Spending Score (1-100)']
for i in columns:
    plt.figure()
    sns.kdeplot(df[i],shade=True,hue=df['Gender'])
```

```python
columns = ['Age', 'Annual Income (k$)','Spending Score (1-100)']
for i in columns:
    plt.figure()
    sns.boxplot(data=df,x='Gender',y=df[i])
    
```

## Bivariate Analysis


```python
fig,ax = plt.subplots(figsize=(10,8))
sns.scatterplot(data=df, x='Annual Income (k$)',y='Spending Score (1-100)' );
```
